## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

## Import data

```python
data= pd.read_csv('airline-passenger-traffic+(1).csv',header=None)
data.columns=['Month','Passengers']
data['Month']=pd.to_datetime(data['Month'],format='%Y-%m')
```

```python
data=data.set_index('Month')
```

```python
data.head(10)
```

## Time Series

### Plot the data

```python
data.plot(figsize=(8,4))
plt.show()
```

```python
data=data.assign(passengers_imputed=data.Passengers.interpolate(method='linear'))
```

```python
data.plot(figsize=(8,4))
plt.show()
```

## Additive seasonality

```python
from pylab import rcParams
import statsmodels.api as sm
rcParams['figure.figsize']=12,8
decomposition=sm.tsa.seasonal_decompose(data.passengers_imputed,model='additive')
fig=decomposition.plot()
plt.show()
```

When the magnitude of the seasonal pattern in the data does not directly correlate with the value of the series, the additive seasonal decomposition may be a better choice to split the time series so that the residual does not have any pattern.

Imagine that you are viewing a time series data of rainfall in a city where there is no difference in the average rainfall received during the last 10 years. In this case, since the average rainfall is the same during the 10 years, the seasonal peaks will almost remain the same height and thus we can use additive decomposition over this data. Additive decomposed time series have a linear variation in the trend.

## Multiplicative Seasonality

```python
from pylab import rcParams
import statsmodels.api as sm
rcParams['figure.figsize']=12,8
decomposition=sm.tsa.seasonal_decompose(data.passengers_imputed,model='multiplicative')
fig=decomposition.plot()
plt.show()
```

When the magnitude of the seasonal pattern in the data increases with an increase in data values and decreases with a decrease in the data values, the multiplicative seasonal decomposition may be a better choice. This means, that for every seasonal pattern in the time series, the distance between the peaks and the troughs keeps increasing

Imagine that you are viewing a time series data of rainfall in a city where now, there is a difference in the average rainfall received during the last 10 years and this keeps increasing every year. In this case, since the average rainfall is increasing during the 10 years, the seasonal peaks will also now keep increasing or the distance between the troughs and the peaks will increase as you progress through the time-series.
